```python
import pandas as pd
data = pd.read_csv('sales records.csv')
print(data.head())
```

```python
filtered_data = data[data['Total'] > 500]
print(filtered_data.head())
sorted_data = data.sort_values(by='Total', ascending=False)
print(sorted_data.head())
```

```python

grouped_data = data.groupby('City')['Total'].sum().reset_index()
print(grouped_data)
```

```python

mean_total = data['Total'].mean()
median_total = data['Total'].median()
std_total = data['Total'].std()

print(f"Mean Total Amount: {mean_total}")
print(f"Median Total Amount: {median_total}")
print(f"Standard Deviation of Total Amount: {std_total}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data['Total'], kde=True)
plt.title('Total Amount Distribution')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='City', y='Total', data=data)
plt.title('Total Amount by City')
plt.xlabel('City')
plt.ylabel('Total Amount')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Total', data=data)
plt.title('Total Amount vs. Quantity')
plt.xlabel('Quantity')
plt.ylabel('Total Amount')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Product line', y='Total', data=data, estimator=sum, ci=None)
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
